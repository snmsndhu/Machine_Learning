# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
 
x = [1, 2, 3, 4 ]
y = [11, 22,33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot data

ax.plot(x , y)

# 4. Customize plot

ax.set(title="Simple Plot",
       xlabel= "x-axis",
       ylabel= "y_axis"
       )

# 5. Save & show (you save the whole figure)
fig.savefig("../03_matplotlib/plots/sample_plot.png")
```

# Making figures with numpy arrays

we want:

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data

x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data nad create a line plot

fig, ax = plt.subplots()
ax.plot(x, x ** 2);
```

```python
# Use the same data and create a Scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another Scatter Plot

fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make the plot from the dictionary

nut_butter_prices = {
    "Almond" : 10,
    "Peanut": 8,
    "Cashew": 12
}
fig, ax = plt.subplots()
x = nut_butter_prices.keys()
y = nut_butter_prices.values()
ax.bar(x, y);
```

```python
# Use some data and create the Histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# There are two options for subplots

# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize = (10, 5))


# Plot to each different axis

ax1.plot(x, x/2)

ax2.scatter(np.random.random(10), np.random.random(10));
```
