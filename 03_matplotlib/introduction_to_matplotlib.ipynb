# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
 
x = [1, 2, 3, 4 ]
y = [11, 22,33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot data

ax.plot(x , y)

# 4. Customize plot

ax.set(title="Simple Plot",
       xlabel= "x-axis",
       ylabel= "y_axis"
       )

# 5. Save & show (you save the whole figure)
fig.savefig("../03_matplotlib/plots/sample_plot.png")
```

# Making figures with numpy arrays

we want:

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# Create some data

x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data nad create a line plot

fig, ax = plt.subplots()
ax.plot(x, x ** 2);
```

```python
# Use the same data and create a Scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another Scatter Plot

fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make the plot from the dictionary

nut_butter_prices = {
    "Almond" : 10,
    "Peanut": 8,
    "Cashew": 12
}
fig, ax = plt.subplots()
x = nut_butter_prices.keys()
y = nut_butter_prices.values()
ax.bar(x, y);
```

```python
# Use some data and create the Histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# There are two options for subplots

# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize = (10, 5))


# Plot to each different axis

ax1.plot(x, x/2)

ax2.scatter(np.random.random(10), np.random.random(10))

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax4.hist(np.random.randn(1000));
```

```python
# Option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize =(10, 5))

# Plot to each different index

ax[0, 0].plot(x, x/2);
```

### Which one should we use ? (pypolt vs matplotlib OO method ?)

* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advanced, use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python

#Pyplot Method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c= "target");
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             ax=ax);

ax.set_xlim([45, 100])
```

```python
## OO method from scratch

fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize the plot

ax.set(title="Heart Disease and cholesterol Levesl",
       xlabel = "Age",
       ylabel = "Cholesterol")

# Add a legend

ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line

ax.axhline(over_50["chol"].mean(),
           linestyle = '--')
```

```python
# Subplot of age, chol, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize = (10, 10))

# Add data to ax0
scatter = ax0.scatter(x= over_50["age"],
                      y= over_50["chol"],
                      c = over_50["target"])

ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel = "Age",
        ylabel = "cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline

ax0.axhline(y= over_50["chol"].mean(), linestyle = "--")

####  ax1 ####

# Add the data in ax1

scatter1 = ax1.scatter(x = over_50["age"],
                       y = over_50["thalach"],
                       c = over_50["target"])

ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")


# Add a legend to ax1

ax1.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline

ax1.axhline(y = over_50["thalach"].mean(), linestyle = "--");
```
