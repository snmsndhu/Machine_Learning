```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)

fig, ax = plt.subplots()
ax.plot(x, y);
```

# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
 
x = [1, 2, 3, 4 ]
y = [11, 22,33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot data

ax.plot(x , y)

# 4. Customize plot

ax.set(title="Simple Plot",
       xlabel= "x-axis",
       ylabel= "y_axis"
       )

# 5. Save & show (you save the whole figure)
```
